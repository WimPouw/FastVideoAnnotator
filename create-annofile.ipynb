```python
import os
import glob


curfolder = os.getcwd()
videofolder = os.path.join(curfolder, "Stimuli")

video_list = glob.glob(os.path.join(videofolder, "*.mp4"))
print(video_list)
```

```python
# Create df with column FILE and list all videos per row

import pandas as pd

df_videos = pd.DataFrame(video_list, columns=["FILE"])
df_videos
```

```python
# save it as csv
df_videos.to_csv(os.path.join(videofolder, "stimuli_annotation.csv"), index=False)
```
